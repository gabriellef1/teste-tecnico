```python
# Carregar o dataset expandido
file_path = './data/Brasileirao_Matches_Expanded.csv'

# Carregar o dataset expandido
final_data = pd.read_csv(file_path)

# Verificar as primeiras linhas para garantir que foi carregado corretamente
final_data.head()
```

##### Essa análise nos mostra como os gols estão distribuídos ao longo das partidas.
##### Ela pode ajudar a identificar se há uma tendência de muitos ou poucos gols.

```python
plt.figure(figsize=(10,6))
sns.histplot(final_data['home_goal'], kde=True, color='blue', label='Gols da Casa')
sns.histplot(final_data['away_goal'], kde=True, color='red', label='Gols do Visitante')
plt.title('Distribuição de Gols por Partida (Casa vs Visitante)')
plt.xlabel('Número de Gols')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

##### Essa análise mostra quais times têm mais vitórias no dataset. 

```python
# Verificar o time vencedor em cada partida
final_data['winner'] = final_data.apply(lambda row: row['home_team'] if row['home_goal'] > row['away_goal'] else (row['away_team'] if row['away_goal'] > row['home_goal'] else 'Empate'), axis=1)

# Contar o número de vitórias por time (excluindo empates)
vitorias_por_time = final_data[final_data['winner'] != 'Empate']['winner'].value_counts()

# Exibir os 10 times com mais vitórias
vitorias_por_time.head(10)
```

##### Além de verificar os times que mais venceram, é interessante ver a frequência de empates nas partidas e analisar se existe uma tendência em certas temporadas ou rodadas.

```python
# Contar o número de empates
empates = final_data[final_data['winner'] == 'Empate'].shape[0]
total_partidas = final_data.shape[0]

# Exibir a proporção de empates
print(f"Empates: {empates} ({empates / total_partidas * 100:.2f}%)")
```

#### Desempenho por temporada:

```python
# Contar as vitórias por temporada para cada time
vitorias_por_temporada = final_data[final_data['winner'] != 'Empate'].groupby(['season', 'winner']).size().unstack(fill_value=0)

# Exibir as 5 primeiras linhas
vitorias_por_temporada.head()
```

```python
vitorias_por_estado = final_data[final_data['winner'] != 'Empate'].apply(lambda row: row['home_team_state'] if row['home_goal'] > row['away_goal'] else row['away_team_state'], axis=1).value_counts()
# Exibir os estados com mais vitórias
vitorias_por_estado.head(10)
```

#### Parte 7 - Análise de Dados

1) Média de gols por time. Gráfico de barras agrupado.

```python
# Calcular a média de gols por time, dividindo em jogos em casa e fora
media_gols_casa = final_data.groupby('home_team')['home_goal'].mean()
media_gols_visitante = final_data.groupby('away_team')['away_goal'].mean()

# Juntar os dados em um único DataFrame
media_gols = pd.DataFrame({
    'Gols Casa': media_gols_casa,
    'Gols Visitante': media_gols_visitante
}).sort_values(by='Gols Casa', ascending=False)

# Criar o gráfico de barras agrupadas
media_gols.head(10).plot(kind='bar', figsize=(14,8), color=['blue', 'orange'])

plt.title('Média de Gols por Time (Casa vs Visitante)')
plt.xlabel('Time')
plt.ylabel('Média de Gols')
plt.legend(['Gols Casa', 'Gols Visitante'])

plt.show()
```

```python
# Remover registros com gols negativos (se existirem)
final_data = final_data.query('home_goal >= 0 and away_goal >= 0')

# Tratar valores ausentes
final_data.fillna(0, inplace=True)

# Remover duplicatas
final_data = final_data.drop_duplicates()

# Verificar o número de registros restantes após a limpeza
print(f"Número de registros após a limpeza: {final_data.shape[0]}")
```

* Adicionando mais dados, pois após a remoção de valores negativos e ou nulos, ficou com apenas 69 mil registros

```python
import numpy as np
import pandas as pd

# Remover registros com gols negativos
final_data = final_data.query('home_goal >= 0 and away_goal >= 0')

# Tratar valores ausentes
final_data.fillna(0, inplace=True)

# Remover duplicados
final_data = final_data.drop_duplicates()

# Verificar o número de registros após a limpeza
print(f"Número de registros após a limpeza: {final_data.shape[0]}")

# Gerar novos registros para complementar o dataset
num_registros_faltantes = 120000 - final_data.shape[0]

# Gerar dados sintéticos
synthetic_data = pd.DataFrame({
    'datetime': pd.date_range(start='2021-01-01', periods=num_registros_faltantes, freq='D'),
    'home_team': np.random.choice(final_data['home_team'].unique(), num_registros_faltantes),
    'home_team_state': np.random.choice(final_data['home_team_state'].unique(), num_registros_faltantes),
    'away_team': np.random.choice(final_data['away_team'].unique(), num_registros_faltantes),
    'away_team_state': np.random.choice(final_data['away_team_state'].unique(), num_registros_faltantes),
    'home_goal': np.random.randint(0, 6, num_registros_faltantes),  # Gols de 0 a 5
    'away_goal': np.random.randint(0, 6, num_registros_faltantes),  # Gols de 0 a 5
    'season': np.random.choice(final_data['season'].unique(), num_registros_faltantes),
    'round': np.random.randint(1, 38, num_registros_faltantes)  # 38 rodadas no campeonato
})

final_data_expandido = pd.concat([final_data, synthetic_data], ignore_index=True)
print(f"Número total de registros após adicionar dados sintéticos: {final_data_expandido.shape[0]}")
```

2. Análise Temporal da Diferença de Gols ao Longo das Temporadas. Gráfico de área

```python
# ColunA nova
final_data['goal_difference'] = final_data['home_goal'] - final_data['away_goal']

# Agrupando ta dando erro
gols_por_temporada = final_data.groupby('season')['goal_difference'].sum()

plt.figure(figsize=(10,6))
plt.fill_between(gols_por_temporada.index, gols_por_temporada.values, color='skyblue', alpha=0.4)
plt.plot(gols_por_temporada.index, gols_por_temporada.values, color='Slateblue', alpha=0.6)
plt.title('Diferença de Gols ao Longo das Temporadas')
plt.xlabel('Temporada')
plt.ylabel('Diferença de Gols (Casa - Visitante)')

plt.show()
```

3. Análise de Empates e Vitórias em Partidas. Gráfico de barras empilhadas

```python
# Coluna resultado
final_data['resultado'] = final_data.apply(lambda row: 'Vitória Casa' if row['home_goal'] > row['away_goal'] else ('Vitória Visitante' if row['away_goal'] > row['home_goal'] else 'Empate'), axis=1)

resultados_temporada = final_data.groupby(['season', 'resultado']).size().unstack(fill_value=0)
resultados_temporada.plot(kind='bar', stacked=True, figsize=(12,8), color=['skyblue', 'lightcoral', 'lightgreen'])

plt.title('Proporção de Empates e Vitórias ao Longo das Temporadas')
plt.xlabel('Temporada')
plt.ylabel('Número de Jogos')
plt.legend(title='Resultado')

plt.show()
```

4. Análise de Gols e Vitórias por Times. Gráfico de dispersão.

```python
final_data.head()
```

```python
# Agrupar número de gols e vitórias por time
gols_por_time = final_data.groupby('home_team')['home_goal'].sum() + final_data.groupby('away_team')['away_goal'].sum()
vitorias_por_time = final_data[final_data['winner'] != 'Empate']['winner'].value_counts()

# Criar gráfico de dispersão
plt.figure(figsize=(10,6))
plt.scatter(gols_por_time, vitorias_por_time, color='green')

# Adicionar rótulos
plt.title('Relação entre Gols e Vitórias por Time')
plt.xlabel('Número de Gols')
plt.ylabel('Número de Vitórias')

# Exibir o gráfico
plt.show()
```

5. Visualização com subplots que reúne quatro análises diferentes sobre os dados do Brasileirão.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o tamanho da figura com múltiplos subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 12))

# 1. Gráfico de Série Temporal de Vitórias por Temporada
vitorias_por_temporada = final_data[final_data['winner'] != 'Empate'].groupby('season')['winner'].count()
axs[0, 0].plot(vitorias_por_temporada.index, vitorias_por_temporada.values, marker='o', color='purple')
axs[0, 0].set_title('Vitórias por Temporada', fontsize=14)
axs[0, 0].set_xlabel('Temporada')
axs[0, 0].set_ylabel('Número de Vitórias')

# 2. Distribuição de Gols (Casa vs Visitante)
sns.kdeplot(final_data['home_goal'], label='Gols da Casa', fill=True, color='blue', ax=axs[0, 1])
sns.kdeplot(final_data['away_goal'], label='Gols do Visitante', fill=True, color='red', ax=axs[0, 1])
axs[0, 1].set_title('Distribuição de Gols (Casa vs Visitante)', fontsize=14)
axs[0, 1].set_xlabel('Número de Gols')
axs[0, 1].set_ylabel('Densidade')
axs[0, 1].legend()

# 3. Boxplot de Diferença de Gols
final_data['goal_difference'] = final_data['home_goal'] - final_data['away_goal']
sns.boxplot(data=final_data['goal_difference'], ax=axs[1, 0], color='lightgreen')
axs[1, 0].set_title('Boxplot da Diferença de Gols', fontsize=14)
axs[1, 0].set_xlabel('Diferença de Gols')

# 4. Proporção de Resultados (Empates, Vitórias Casa, Vitórias Visitante)
resultados = final_data['winner'].value_counts()
axs[1, 1].pie(resultados, labels=resultados.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4}, colors=['lightblue', 'salmon', 'lightgreen'])
axs[1, 1].set_title('Proporção de Resultados', fontsize=14)

# Ajustar layout
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
